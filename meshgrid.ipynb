# 格子点の作成

## import library

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
```

## 2変数関数の定義

\\[ f(z) = x^2 - y^2 \\]

```python
def f(x, y):
    return x**2 - y**2
```

## パラメータ定義

```python
# 配列の要素数
n = 5**2

# 格子点を作成
x = np.linspace(0, 5, n)
y = np.linspace(0, 5, n)
X, Y = np.meshgrid(x, y)

# zの計算
Z = f(X, Y)
```

## 曲面の描画

```python
# Figureと3DAxesを追加
fig = plt.figure(figsize = (8, 6))
ax = fig.add_subplot(111, projection="3d")

# 3DAxesの設定
ax.set_xlabel("x", size = 15)
ax.set_ylabel("y", size = 15)
ax.set_zlabel("z", size = 15)
ax.view_init(45, 45)

# 曲面を描画
ax.plot_surface(X, Y, Z, cmap = "autumn")

plt.show()
```

# 2次元格子点

## import library

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
```

## 2次元格子点classの定義

```python
class Mesh2D:
    def __init__(self, xrange, yrange, n):
        self.n = n
        self.x_min = xrange[0]
        self.x_max = xrange[1]
        self.y_min = yrange[0]
        self.y_max = yrange[1]
        self.x = np.linspace(xrange[0], xrange[1], n)
        self.y = np.linspace(yrange[0], yrange[1], n)
        self.X, self.Y = np.meshgrid(self.x, self.y)
```

## 2変数関数の定義

```python
def f(x, y):
    return np.cos(x + y)
```

## 曲面描画

```python
# Mesh2Dクラスのインスタンスを生成
m = Mesh2D([-4,4], [-4, 4], 65)

# 高度式を計算
Z = f(m.X, m.Y)

# Figureと3DAxesを追加
fig = plt.figure(figsize = (8, 6))
ax = fig.add_subplot(111, projection="3d")

# 3DAxesの設定
ax.set_xlabel("x", size = 15)
ax.set_ylabel("y", size = 15)
ax.set_zlabel("z", size = 15)
ax.view_init(60, 30)

# 曲面を描画
ax.plot_surface(m.X, m.Y, Z, cmap = "winter")

plt.show()
```
