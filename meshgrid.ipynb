# 格子点の作成

## import library

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
```

## 2変数関数の定義

```python
def f(x, y):
    return x**2 - y**2
```

## パラメータ定義

```python
# 配列の要素数
n = 9

# 格子点を作成
x = np.linspace(0, 5, n)
y = np.linspace(0, 5, n)
X, Y = np.meshgrid(x, y)

# zの計算
Z = f(X, Y)
```

## 曲面の描画

```python
# Figureと3DAxesを追加
fig = plt.figure(figsize = (8, 6))
ax = fig.add_subplot(111, projection="3d")

# 3DAxesの設定
ax.set_xlabel("x", size = 15)
ax.set_ylabel("y", size = 15)
ax.set_zlabel("z", size = 15)
ax.view_init(45, 45)

# 曲面を描画
ax.plot_surface(X, Y, Z, cmap = "autumn")

plt.show()
```
